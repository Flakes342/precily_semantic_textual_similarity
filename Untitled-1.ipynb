```python
# This is a problem in Natural Language Processing (NLP), and we must do text embedding 
# before developing any deep learning model in NLP. Text embedding is a method of converting text into numerical vectors.
# We can detect how similar two texts are by determining how close these vectors using 
# euclidean distance or cosine similarity approach after transforming the words into vectors.
# We employed Google's Universal Sentence Encoder (USE), 
# which converts text into higher-dimensional vectors for our semantic similarity test.


# importing required libraries
import os
import numpy as np                 # linear algebra
import pandas as pd                # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf            # tensorflow for ML models
import tensorflow_hub as hub       # contains USE4
from numpy import dot              # to calculate the dot product of two vectors
from numpy.linalg import norm      # for finding the norm of a vector

module_import = "https://tfhub.dev/google/universal-sentence-encoder/4" #Encoder is imported from this URL
encoder = hub.load(module_import)


def embed(input):                                               # initialising an embed function which returns the encoded text
  return encoder(input)

# Importing training dataset
df = pd.read_csv("C:/Users/DELL/OneDrive\Desktop/Precily_Task/Precily_Text_Similarity.csv")

sim_score = []                                                  # This list will contain the cosin similarity value for each vector pair present.
for i in range(len(df)):
  messages = [df['text1'][i], df['text2'][i]]                   # storing each text pair in messages
  message_embeddings = embed(messages)                          # converting the sentence pair to vector pair using the embed() function
  t = tf.make_ndarray(tf.make_tensor_proto(message_embeddings)) # converting the vector in the form of numpy array and making a numpy array of all the nparrays
  cos_sim = dot(t[0], t[1])/(norm(t[0])*norm(t[1]))             # cosine between the two word-vectors
  sim_score.append(cos_sim)                                     # Appending the values into the sim_score list

  
Sim_score = pd.DataFrame(sim_score, columns = ['Similarity_Score'])           # converting the ans list into Dataframe so that we can add it to our original dataframe
df = df.join(Sim_score)                                                       # Joining the Similarity_Score Dataframe (Sim_score) to our main Dataframe
df['Similarity_Score'] = df['Similarity_Score'] + 1                           # adding 1 to each of the values of Similarity_Score to make the values from 0 to 2. (Initially it was from [-1,1])
df['Similarity_Score'] = df['Similarity_Score']/df['Similarity_Score'].max()  # Normalizing the Similarity_Score to get the value between 0 and 1 using max value as 1.


submission = df[['Similarity_Score']]                           # creating new submission dataframe with the Similarity score column
submission.to_csv('mycsvfile.csv',index=True)
sub = pd.read_csv("mycsvfile.csv")


class model:
  def __init__ (self):
    self = None

  def predict (self, text1, text2):
    messages = [text1, text2]
    message_embeddings = embed(messages)
    t = tf.make_ndarray(tf.make_tensor_proto(message_embeddings))
    cos_sim = dot(t[0], t[1])/(norm(t[0])*norm(t[1]))
    y = (cos_sim+1)/2
    return (y)
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

import pickle
with open('precily_task.pkl', 'wb') as f:
  pickle.dump(model, f)
with open('precily_task.pkl', 'rb') as f:
    model_loaded = pickle.load(f)

sub.head()
```

```python
import numpy as np                 # linear algebra
import pandas as pd                # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf            # tensorflow for ML models
import tensorflow_hub as hub       # contains USE4
from numpy import dot              # to calculate the dot product of two vectors
from numpy.linalg import norm      # for finding the norm of a vector
import spacy_universal_sentence_encoder


class MachineLearningCode:
  
  def __init__ (self):
    self.text1 = None
    self.text2 = None
    
   
  


  def embed(self, input):
    #module_import = "https://tfhub.dev/google/universal-sentence-encoder/4" #Encoder is imported from this URL
    encoder = spacy_universal_sentence_encoder.load_model('xx_use_lg')# initialising an embed function which returns the encoded text
    return encoder(input)

  def predict (self, text1, text2):
    messages = [text1, text2]
    message_embeddings = self.embed(messages)
    t = tf.make_ndarray(tf.make_tensor_proto(message_embeddings))
    cos_sim = dot(t[0], t[1])/(norm(t[0])*norm(t[1]))
    y = (cos_sim+1)/2
    return (round(y),5)
```
